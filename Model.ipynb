```python
import pandas as pd
import re
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
def clean_skills(skill_str):
    if pd.isna(skill_str) or skill_str.strip() == "":
        return []
    skill_str = re.sub(r'[^a-zA-Z, ]', '', skill_str)  # remove special chars/numbers
    skills = [s.strip().lower() for s in skill_str.split(",") if s.strip() != ""]
    return skills
```

```python
# Sample candidate data
candidates_data = pd.DataFrame(filtered_df)

candidates_data["linkedinSkillsLabel"] = candidates_data["linkedinSkillsLabel"].apply(clean_skills)
```

```python
job_skills_required = {
  "must": [
    "python",
    "sql",
    "machine learning",
    "statistics",
    "pandas",
    "scikit-learn"
  ],
  "nice": [
    "numpy",
    "deep learning",
    "nlp",
    "transformers",
    "data structures",
    "algorithms",
    "git"
  ],
  "bonus": [
    "pytorch",
    "tensorflow",
    "mlops",
    "docker",
    "langchain",
    "aws",
    "gcp",
    "azure"
  ]
}
```

```python
# Create skill vocabulary
all_skills = set()
for s in candidates_data["linkedinSkillsLabel"]:
    all_skills.update(s)
for k in job_skills_required.values():
    all_skills.update(k)

all_skills = sorted(list(all_skills))
```

```python
# ---------------- Step 3: Create Vectors ----------------
def create_vector(skills, vocab, weights=None):
    vec = []
    for v in vocab:
        if v in skills:
            if weights and v in weights:
                vec.append(weights[v])
            else:
                vec.append(1)
        else:
            vec.append(0)
    return np.array(vec)
```

```python
# Assign weights
weights = {}
for s in job_skills_required["must"]: weights[s] = 3
for s in job_skills_required["nice"]: weights[s] = 2
for s in job_skills_required["bonus"]: weights[s] = 1
    
job_vec = create_vector(all_skills, all_skills, weights)
```

```python
# ---------------- Step 4: Score Candidates ----------------
results = []
for idx, row in candidates_data.iterrows():
    cand_vec = create_vector(row["linkedinSkillsLabel"], all_skills)
    
    # Cosine similarity
    sim = cosine_similarity([cand_vec], [job_vec])[0][0]
    
    # Missing skills
    missing = len(set(job_skills_required["must"] + 
                      job_skills_required["nice"] + 
                      job_skills_required["bonus"]) - set(row["linkedinSkillsLabel"]))
    
    score = max(0, 100 - (missing * 10))
    
    results.append([row["fullName"], row["linkedinSkillsLabel"], score, sim])

results_df = pd.DataFrame(results, columns=["fullName", "linkedinSkillsLabel", "Score", "CosineSim"])
results_df = results_df.sort_values(by="Score", ascending=False)

print(results_df)
```

```python
# 1. Data Cleaning

def clean_text(text):
    if pd.isnull(text):
        return ""
    text = text.lower()  # lowercase
    text = ''.join([c if c.isalpha() or c.isspace() else ' ' for c in text])  # remove special chars/numbers
    return text

df_new['linkedinSkillsLabel'] = df_new['linkedinSkillsLabel'].apply(clean_text)
jobs['job_skills_required'] = jobs['job_skills_required'].apply(clean_text)
```

```python
df_new.head()
```

```python

# 2. TF-IDF Vectorization

vectorizer = TfidfVectorizer()
```

```python
# Combine all text (candidates + job) so TF-IDF has same vocabulary
all_text = df_new['linkedinSkillsLabel'].tolist() + jobs['job_skills_required'].tolist()

tfidf_matrix = vectorizer.fit_transform(all_text)
```

```python
# Split back into candidate vectors & job vectors
candidate_tfidf = tfidf_matrix[:len(df_new)]
job_tfidf = tfidf_matrix[len(df_new):]
```

```python
# 3. Cosine Similarity

similarity_scores = cosine_similarity(candidate_tfidf, job_tfidf)
```

```python
# 4: Add scores back to candidates

df_new['similarity_score'] = similarity_scores.flatten() * 100  # scale to percentage

```
